```python
# Dependencies and Setup
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
import requests
import time
import openweathermapy.core as owm

# Import API key
from api_keys import api_key

# Incorporated citipy to determine city based on latitude and longitude
from citipy import citipy

# Output File (CSV)
output_data_file = "./output_data/cities.csv"

# Range of latitudes and longitudes
lat_range = (-90, 90)
lng_range = (-180, 180)
```

```python
# List for holding lat_lngs and cities
lat_lngs = []
cities = []

# Create a set of random lat and lng combinations
lats = np.random.uniform(low=-90.000, high=90.000, size=1500)
lngs = np.random.uniform(low=-180.000, high=180.000, size=1500)
lat_lngs = zip(lats, lngs)

# Identify nearest city for each lat, lng combination
for lat_lng in lat_lngs:
    city = citipy.nearest_city(lat_lng[0], lat_lng[1]).city_name
    
    # If the city is unique, then add it to a our cities list
    if city not in cities:
        cities.append(city)

# Print the city count to confirm sufficient count
len(cities)
```

```python
settings = {"units":"metric","appid":api_key}

weather_data = []
cities_found = []
counter = 0
set_c = 1

print("Begginig Data Retreival")
print("-----------------------------")

for city_w in cities:
    try:
        weather_data.append(owm.get_current(city_w, **settings))
        
        if counter <= 50:
            counter +=1
        else:
            counter = 1
            set_c +=1
            
        print(f"Processing Record {counter} of Set {set_c} | {city_w}")
        cities_found.append(city_w)
        
    except:
        print("City not found. Skipping...")
```

```python
# Create an "extracts" object to get the temperature, latitude, and longitude in each city
summary = ["clouds.all", "sys.country", "dt", "main.humidity", "coord.lat", "coord.lon", "main.temp_max", "wind.speed"]

# Create a Pandas DataFrame with the results
data = [response(*summary) for response in weather_data]

column_names = ["Cloudiness", "Country", "Date", "Humidity", "Lat", "Lng", "Max Temp", "Wind Speed"]

weather_df = pd.DataFrame(data, index=cities_found, columns=column_names).reset_index()

weather_df.rename(columns={"index": "City"}, inplace=True)

weather_df.to_csv(output_data_file)

weather_df.head()
```

#### Latitude vs. Temperature Plot

```python
today = (time.strftime("%m/%d/%Y"))

# Tell matplotlib to create a scatter plot based upon the above data
plt.scatter(weather_df["Lat"], weather_df["Max Temp"], marker="o", facecolors="blue",edgecolors="black")
plt.ylim(-30, 110)
plt.xlim(-60,90)

plt.title(f"City Latitude vs. Max Temperature ({today})")
plt.xlabel("Latitude")
plt.ylabel("Max Temperature (F)")

plt.savefig("./LatVsMaxTemp.png")
```
